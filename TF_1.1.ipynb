### Importing required packages

```python
import tensorflow as tf
import pandas as pd
import numpy as np
from tensorflow import keras
```

### Creating the feedin values in numpy array 

```python
x_values = np.array([-1.0, 0.0, 2.0, 3.0, 4.0], dtype=float)
y_values = np.array([-3.0, -1.0, 3.0, 5.0, 7.0], dtype=float)
```

### Creating the simple model

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

### Define the optimizer and loss functions for our model

```python
model.compile(optimizer='sgd',loss='mean_squared_error')
```

### Training the Neural Network

```python
model.fit(x_values,y_values, epochs=500)
```

#### So we have trianed out model to learn relationship btw X and Y
#### Lets checkout what our model has predicted

```python
print(model.predict([10.0]))
```

### As we can see that the output is very very close to 19
### We got little under 19 because NN deals with probabilities; with the data that we fed to NN it calcualted very high probability relationship between X and Y
